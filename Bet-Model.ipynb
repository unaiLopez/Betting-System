```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
import numpy as np
from sklearn.metrics import roc_curve, roc_auc_score, auc
from sklearn.preprocessing import label_binarize
from sklearn.naive_bayes import GaussianNB
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X.astype(np.float64))

#pca = PCA(n_components=30)
#principal_components = pca.fit_transform(X_scaled)

X_train, Y_train, X_test, Y_test = X_scaled[:-1180], Y[:-1180], X_scaled[-1180:], Y[-1180:]
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
def multiclass_roc_auc_score(model, X_scaled, Y, num_matches, average="macro"):

    # Binarize the output
    y = label_binarize(Y, classes=[1, 2, 3])
    n_classes = y.shape[1]

    # shuffle and split training and test sets
    X_train, y_train, X_test, y_test = X_scaled[:-num_matches], y[:-num_matches], X_scaled[-num_matches:], y[-num_matches:]

    # Learn to predict each class against the other
    try:
        y_score = OneVsRestClassifier(model, n_jobs=-1).fit(X_train, y_train).decision_function(X_test)
    except:
        y_score = OneVsRestClassifier(model, n_jobs=-1).fit(X_train, y_train).predict_proba(X_test)
    

    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Compute micro-average ROC curve and ROC area
    fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_score.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
    
    plt.figure()
    lw = 2
    plt.plot(fpr[2], tpr[2], color='darkorange', lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[2])
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()
```

```python
def train_analize_model(model, X_train, Y_train, X_test, Y_test, bet_money, matches, num_matches):
    last_matches = matches.tail(num_matches)
    last_matches.index = range(num_matches)
    profit_bet365 = bet_money*(-num_matches)
    profit_betwin = bet_money*(-num_matches)
    profit_interwetten = bet_money*(-num_matches)
    profit_william_hill = bet_money*(-num_matches)
    profit_vc_bet = bet_money*(-num_matches)
    multiclass_roc_auc_score(model, X_scaled, Y, num_matches)
    import warnings
    warnings.filterwarnings('ignore')
    model.fit(X_train, Y_train)
    prediction = model.predict(X_test)
    scores_model_accuracy = cross_val_score(model, X_scaled, Y, scoring="accuracy", cv=10)
    print("Accuracy: ", precision_score(Y_test, prediction, average="micro"))
    print("Recall: ", recall_score(Y_test, prediction, average="micro"))
    print("F1: ", f1_score(Y_test, prediction, average="micro"))
    print(scores_model_accuracy)
    print("Mean Accuracy Cross Validation 10 Folds: ", scores_model_accuracy.mean())
    print()
    for i in range(num_matches):
        print("Prediction of Match "+str(i)+": "+str(prediction[i]==Y_test[i])+" ||| Match Result: "+str(prediction[i]))
        if prediction[i] == Y_test[i]:
            if prediction[i] == 1:
                profit_bet365 += last_matches.loc[i, "Bet365 Home Win Odds"]*bet_money
                profit_betwin += last_matches.loc[i, "Bet&Win Home Win Odds"]*bet_money
                profit_interwetten += last_matches.loc[i, "Interwetten Home Win Odds"]*bet_money
                profit_william_hill += last_matches.loc[i, "William Hill Home Win Odds"]*bet_money
                profit_vc_bet += last_matches.loc[i, "VC Bet Home Win Odds"]*bet_money
            elif prediction[i] == 2:
                profit_bet365 += last_matches.loc[i, "Bet365 Draw Odds"]*bet_money
                profit_betwin += last_matches.loc[i, "Bet&Win Draw Odds"]*bet_money
                profit_interwetten += last_matches.loc[i, "Interwetten Draw Odds"]*bet_money
                profit_william_hill += last_matches.loc[i, "William Hill Draw Odds"]*bet_money
                profit_vc_bet += last_matches.loc[i, "VC Bet Draw Odds"]*bet_money
            elif prediction[i] == 3:
                profit_bet365 += last_matches.loc[i, "Bet365 Away Win Odds"]*bet_money
                profit_betwin += last_matches.loc[i, "Bet&Win Away Win Odds"]*bet_money
                profit_interwetten += last_matches.loc[i, "Interwetten Away Win Odds"]*bet_money
                profit_william_hill += last_matches.loc[i, "William Hill Away Win Odds"]*bet_money
                profit_vc_bet += last_matches.loc[i, "VC Bet Away Win Odds"]*bet_money
                
    print()
    print("Profit in Bet365 Betting in all Matches: "+str(profit_bet365)+" euros")
    print("Profit in Bet&Win Betting in all Matches: "+str(profit_betwin)+" euros")
    print("Profit in Interwetten Betting in all Matches: "+str(profit_interwetten)+" euros")
    print("Profit in William Hill Betting in all Matches: "+str(profit_william_hill)+" euros")
    print("Profit in VCBet Betting in all Matches: "+str(profit_vc_bet)+" euros")
                
```
